```python
#Bibliotecas
import geopandas as gpd
import mapclassify
import matplotlib.pyplot as plt
```

```python
#Leitura do arquivo
```

```python
gdf = gpd.read_file('/Teste_Geo.shp')
```

```python
gdf.head()
```

```python
#Quantis 5 classes
gdf.plot(column='IVS',scheme='Quantiles',k=5,figsize=(12,8),legend=True)
```

```python
#Quebras naturais 5 classes
gdf.plot(column='IVS',scheme='naturalbreaks',k=5,figsize=(12,8),legend=True)
```

```python
#Intervalos iguais 5 classes
gdf.plot(column='IVS',scheme='equalinterval',k=5,figsize=(12,8),legend=True)
```

```python
#Visualizando classes diretamente
q5=mapclassify.Quantiles(gdf['IVS'],k=5)
```
